There are 3 different ways to perform string formatting:-

- Formatting with % Operator.
- Formatting with format() string method.
- Formatting with string literals, called f-strings.

```python
x = "Hello"
print("This is a great word",x)
```

Formatting with % Operator.

```python
x = 95
print("I scored %s marks in my 12th class Chemistry" %x)
```

```python
x = 7
print("This is %fth class of Python." %x)
```

```python
x = 7
y = 10
print("Sum of %s and %s is %s" %(x,y, x+y) +"%.")
```

```python
x=6
y=8
print("sum of %f and %f is %s" %(x,y, x+y))
```

‘%s’ is used to inject strings similarly ‘%d’ for integers, ‘%f’ for floating-point values, ‘%b’ for binary format.

Format work by putting in one or more replacement fields and placeholders defined by a pair of curly braces { } into a string and calling the str.format().

```python
# Syntax: ‘String here {} then also {}’.format(‘something1′,’something2’)
```

```python
print('String here {} then also {}'.format('something1','something2'))
```

```python
x = "equal"
print('We all are {} {}.'.format(x, x))
```

```python
x = "equal"
print('We all are {t} {t}.'.format(t = x))
```

```python
print('{2} {1} {0}'.format('directions','the', 'Read'))
```

```python
print('a: {a}, b: {b}, c: {c}'.format(a = 1, b = 'Two', c = 12.3))
```

```python
print('The {z} {p} was alright, but the {p} {p} was tough.'.format(z = "first",p = 'second'))
```
