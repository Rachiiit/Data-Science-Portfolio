```python
first = pd.DataFrame({ 'id':[1,2,3,4,5], 
                     'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
                     'subject_id':['sub1','sub2','sub4','sub6','sub5']}) 
first
```

```python
second = pd.DataFrame( {'id':[1,2,3,4,5], 
                       'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
                       'subject_id':['sub2','sub4','sub3','sub6','sub5']}) 
second
```

```python
lrid = pd.merge(first, second, on = 'id')
lrid
```

```python
lrsub = pd.merge(first, second, on = 'subject_id')    #default how value is inner i.e. intersection
lrsub
```

```python
lrsubOuter = pd.merge(first, second, on = 'subject_id', how = 'outer')    #outer means union
lrsubOuter
```

```python
lrsubOuter = pd.merge(first, second, on = 'id', how = 'outer')    #outer means union
lrsubOuter
```

```python
lrsubLeft = pd.merge(first, second, on = 'subject_id', how = 'left')    
#left means all the values from left but common values from the right dataframe
lrsubLeft
```

```python
lrsubRight = pd.merge(first, second, on = 'subject_id', how = 'right')
lrsubRight
```
