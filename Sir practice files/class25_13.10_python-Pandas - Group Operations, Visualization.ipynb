```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
plt.plot(x,y)
plt.xlabel('Angle')
plt.ylabel('Sine of Angle')
plt.title('Sine Wave')
# plt.show()
```

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.sin(x)
fig = plt.figure()   #Creates an empty figure
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Sine Wave')
ax.set_xlabel('Angle')
ax.set_ylabel('Sine of Angle')
```

```python
x = np.arange(0, math.pi*2, 0.05)
y = np.random.random(126)
fig = plt.figure()   #Creates an empty figure
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Random Wave')
ax.set_xlabel('Random Values')
ax.set_ylabel('Something')
```

```python
y = [1,4,9,16,25,36,49,64]
x1 = [1,16,30,42,55,68,77,88]
x2 = [1,6,12,18,28,40,52,65]
fig = plt.figure()   #Creates an empty figure
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1, y, 'bs--')
l2 = ax.plot(x2, y, 'ro-')
ax.legend(labels = ['TV', 'Smartphone'])    #loc = 'upper left'
ax.set_title('Effect of Advertisement on sales')
ax.set_xlabel('Medium')
ax.set_ylabel('Sales')
```

```python
x = rc['Age']
y = rc['Salary']
fig = plt.figure()   #Creates an empty figure
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y)
ax.set_title('Relation between Age and Salary of NBA Players')
ax.set_xlabel('Age')
ax.set_ylabel('Salary')
```

```python
fig, a = plt.subplots(2,2)
x = np.arange(1,5)
a[0][0].plot(x, x*x)
a[0][0].set_title('Square')
a[0][1].plot(x, np.sqrt(x))
a[0][1].set_title('Square Root')
a[1][0].plot(x, np.exp(x))
a[1][0].set_title('Exponential')
a[1][1].plot(x, np.log10(x))
a[1][1].set_title('Logrithmic')
```

```python
fig = plt.figure()   #Creates an empty figure
ax = fig.add_axes([0,0,1,1])
l = ['C', 'C++', 'Java', 'Python', 'PHP']
s = [23, 17, 35, 29, 12]
ax.bar(l, s, color = 'green', width = .8)
```

```python
d = [[30,25,50,20],
    [40,23,51,17],
    [35,22,45,19]]
x = np.arange(4)     #([0,1,2,3])
fig = plt.figure()   
ax = fig.add_axes([0,0,1,1])
ax.bar(x + 0.00, d[0], color = 'b', width = 0.25)   # x = [0,1,2,3], y = [30,25,50,20] 
ax.bar(x + 0.25, d[1], color = 'g', width = 0.25)   # x = [0.25,1.25,2.25,3.25], y = [40,23,51,17]
ax.bar(x + 0.50, d[2], color = 'r', width = 0.25)   # x = [0.5,1.5,2.5,3.5], y = [35,22,45,19]
```
